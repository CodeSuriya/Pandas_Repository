```python
#convert the float values into text values

def float_to_text(value):
    if isinstance(value, float):
        if value <= 30.0:
            return 'Needs Significant Improvement'
        elif value <= 60.0:
            return 'Meets Expectations'
        else:
            return 'Exceeds Expectations'
    return value  
```

```python
df['Remarks']=df['Remarks'].apply(float_to_text)
```
